#Удаляем признаки с высокой корреляцией не трогая симптомы


```python
columns_to_keep = [
    'нестабильность 0 - нет 1 - крайн или част 2 - полная',
    'Из них полная нестабильность (CG-2)',
    'Реоперация по поводу нестабильности',
    'Операция из-за нестабильности + перелом конструкции',
    'перелом металла',
    'резорбция вокруг винтов',
    'смежный уровень',
    'Severity'
]
```

```python
threshold=0.8
```

```python
# Создаем копию DataFrame без столбцов, которые нужно сохранить
df_temp = df.drop(columns=columns_to_keep, errors='ignore')

    # Вычисляем корреляционную матрицу
corr_matrix = df_temp.corr().abs()

    # Создаем маску для верхнего треугольника
upper_triangle = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)

    # Находим пары столбцов с корреляцией выше порога
high_corr_pairs = np.where((corr_matrix > threshold) & upper_triangle)

    # Собираем столбцы для удаления
columns_to_drop = set()
for i, j in zip(*high_corr_pairs):
        col_i = corr_matrix.columns[i]
        col_j = corr_matrix.columns[j]
        # Сохраняем тот столбец, который имеет меньшую корреляцию с целевой переменной
        # или просто первый, если целевой переменной нет
        columns_to_drop.add(col_j)

    # Удаляем столбцы из временного DataFrame
df_temp = df_temp.drop(columns=columns_to_drop)

    # Возвращаем исходный DataFrame с удаленными столбцами + сохраненные столбцы
columns_to_remove_from_original = set(df.columns) - set(columns_to_keep) - set(df_temp.columns)
df_cleaned = df.drop(columns=columns_to_remove_from_original)

print(f"Удалено столбцов: {len(columns_to_remove_from_original)}")
print(f"Сохранено столбцов: {len(df_cleaned.columns)}")
print(f"Удаленные столбцы: {list(columns_to_remove_from_original)}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import plot_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(f"Размерность данных для нейросети:")
print(f"Размерность данных для нейросети:")
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")

print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

```python
# Преобразуем в numpy arrays
X_train_nn = X_train.values.astype('float32')
X_test_nn = X_test.values.astype('float32')
y_train_nn = y_train.values.astype('float32')
y_test_nn = y_test.values.astype('float32')
```

```python
# Масштабирование данных (важно для нейросетей!)
from sklearn.preprocessing import StandardScaler

scaler_nn = StandardScaler()
X_train_scaled = scaler_nn.fit_transform(X_train_nn)
X_test_scaled = scaler_nn.transform(X_test_nn)

print(f"\nДанные после масштабирования:")
print(f"X_train_scaled: {X_train_scaled.shape}")
print(f"X_test_scaled: {X_test_scaled.shape}")
```

```python
def create_deep_learning_model(input_dim, output_dim):

    model = Sequential([

        Dense(128, activation='relu', input_shape=(input_dim,),
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.5),


        Dense(64, activation='relu',
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.4),

        Dense(32, activation='relu',
              kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),


        Dense(output_dim, activation='sigmoid')
    ])

    # Компиляция модели
    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall']
    )

    return model

```

```python
# Создаем модель
input_dim = X_train_scaled.shape[1]
output_dim = y_train_nn.shape[1]

model = create_deep_learning_model(input_dim, output_dim)

print("Архитектура нейросети:")
model.summary()
```

```python


# Callbacks для улучшения обучения
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=25,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=15,
    min_lr=0.00001,
    verbose=1
)



history = model.fit(
    X_train_scaled, y_train_nn,
    epochs=200,
    batch_size=32,
    validation_data=(X_test_scaled, y_test_nn),
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)
```

```python
def plot_training_history(history):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))

    # Loss
    axes[0, 0].plot(history.history['loss'], label='Training Loss')
    axes[0, 0].plot(history.history['val_loss'], label='Validation Loss')
    axes[0, 0].set_title('Loss')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Accuracy
    axes[0, 1].plot(history.history['accuracy'], label='Training Accuracy')
    axes[0, 1].plot(history.history['val_accuracy'], label='Validation Accuracy')
    axes[0, 1].set_title('Accuracy')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # Precision
    axes[1, 0].plot(history.history['precision'], label='Training Precision')
    axes[1, 0].plot(history.history['val_precision'], label='Validation Precision')
    axes[1, 0].set_title('Precision')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Precision')
    axes[1, 0].legend()
    axes[1, 0].grid(True)

    # Recall
    axes[1, 1].plot(history.history['recall'], label='Training Recall')
    axes[1, 1].plot(history.history['val_recall'], label='Validation Recall')
    axes[1, 1].set_title('Recall')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Recall')
    axes[1, 1].legend()
    axes[1, 1].grid(True)

    plt.tight_layout()
    plt.show()

plot_training_history(history)

```

```python
def create_deep_learning_model(input_dim, output_dim):

    model = Sequential([

        Dense(128, activation='relu', input_shape=(input_dim,)),
        BatchNormalization(),
        Dropout(0.1),


        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.1),

        Dense(32, activation='relu'),
        BatchNormalization(),
        Dropout(0.1),


        Dense(output_dim, activation='sigmoid')
    ])

    # Компиляция модели
    model.compile(
        optimizer=Adam(learning_rate=0.0005),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

```

```python
# Создаем модель
input_dim = X_train_scaled.shape[1]
output_dim = y_train_nn.shape[1]

model = create_deep_learning_model(input_dim, output_dim)

print("Архитектура нейросети:")
model.summary()
```

```python


# Callbacks для улучшения обучения
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=25,
    restore_best_weights=True,
    verbose=1
)

history = model.fit(
    X_train_scaled, y_train_nn,
    epochs=200,
    batch_size=32,
    validation_data=(X_test_scaled, y_test_nn),
    callbacks=[early_stopping],
    verbose=1
)
```

```python
def plot_training_history(history):
    fig, axes = plt.subplots( 2, figsize=(15, 10))

    # Loss
    axes[ 0].plot(history.history['loss'], label='Training Loss')
    axes[ 0].plot(history.history['val_loss'], label='Validation Loss')
    axes[ 0].set_title('Loss')
    axes[ 0].set_xlabel('Epoch')
    axes[ 0].set_ylabel('Loss')
    axes[ 0].legend()
    axes[ 0].grid(True)

    # Accuracy
    axes[ 1].plot(history.history['accuracy'], label='Training Accuracy')
    axes[ 1].plot(history.history['val_accuracy'], label='Validation Accuracy')
    axes[ 1].set_title('Accuracy')
    axes[ 1].set_xlabel('Epoch')
    axes[ 1].set_ylabel('Accuracy')
    axes[1].legend()
    axes[ 1].grid(True)



    plt.tight_layout()
    plt.show()

plot_training_history(history)

```

# Тестируем модель и получаем относительно неплохие результаты

```python
results = model.evaluate(X_test_scaled, y_test_nn, verbose=0)


print(f"Loss: {results[0]:.4f}")
print(f"Accuracy: {results[1]:.4f}")
```
